```python
import pandas as pd
```

```python
df_Sept_5 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200905.txt')
df_Sept_12 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200912.txt')
df_Sept_19 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200919.txt')
```

```python
df_Sept_19.head()
```

```python
df_Sept_12.head()
```

```python
df = pd.concat([df_Sept_5, df_Sept_12, df_Sept_19])
```

```python
df
```

```python
df.info()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['DATE'].head()
```

```python
df.head()
```

```python
df['TIME'] = pd.to_datetime(df['TIME'])
df['TIME'].head()
```

```python
df.head()
```

```python
df['DateTime'] = pd.to_datetime(df.DATE.astype(str) + " " + df.TIME.astype(str))
```
