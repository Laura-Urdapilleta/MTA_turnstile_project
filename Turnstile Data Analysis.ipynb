```python
import pandas as pd
```

```python
#to load the data back in use
df_full = pd.read_pickle('original_turnstile_df.pkl')
```

```python
df_full.head()
```

```python
#create a new column with using the date & time columns,
# then turning it into a datetime object
df_full['DateTime'] = pd.to_datetime(df_full.DATE.astype(str) + " " + df_full.TIME.astype(str))
```

```python
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', None)
```

```python
#grouping by turnstiles locations and summing entries/exits by date
# note that SCP is the only location specific column that changes
turnstiles_df = df_full.groupby(['C/A', 'UNIT', 'STATION', 'SCP', 'DATE' ]).sum()
```

```python
turnstiles_df
```

```python
#this is a single turnstile at one location
first_turnstile = turnstiles_df.loc[lambda turnstiles_df: turnstiles_df['STATION'] == '59 ST'].loc[lambda turnstiles_df: turnstiles_df['SCP'] == '02-00-00'].loc[lambda turnstiles_df: turnstiles_df['UNIT'] == 'R051'].loc[lambda turnstiles_df: turnstiles_df['C/A'] == 'A002']
```
